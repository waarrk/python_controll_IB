## グラフのプロット
matplotlib というライブラリを使う。gnuplotに近い感覚で使える。

```python
# ライブラリのインポート
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 0 ~ 2pi の区間を20点に等分割した等差数列のベクトル
x = np.linspace(0, 2*np.pi , 20)
```

```python
# sin(x) の値のベクトル
y1 = np.sin(x)
```

## グラフをプロットする
plt.plot, plt.show()

```python
plt.plot(x, y1)
plt.show()
```

## 軸ラベルをつける
plt.xlabel(), plt.ylabel()

```python
plt.plot(x, y1)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

## 複数のグラフを重ねる
plt.show() の前にplt.plot()を複数回入れる

```python
y2 = np.cos(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```

## 凡例をつける
 - plt.plot()に引数でlabel = "LABEL"を追加
 - plt.legend()

```python
plt.plot(x, y1,label="sin")
plt.plot(x,y2, label="cos")
plt.legend()
plt.show()
```

## グリッド線をつける
plt.grid()

```python
plt.plot(x, y1)
plt.grid()
plt.show()
```

## 線の色を変える
plt.plot()に引数color="COLOR"を追加

```python
plt.plot(x, y1, color="red")
plt.plot(x, y2, color="blue")
plt.show()
```

## 線の種類を変える
plt.plot()に引数linestyle="LINESTYLE"を追加

```python
plt.plot(x, y1, linestyle="--")
plt.plot(x, y2, linestyle="-.")
plt.show()
```

## Python-control を使ったステップ応答の描画
伝達関数にはtf関数wp使う。例えば、

$$
P(s) = \frac{3}{2s+1}
$$

を定義するなら次のようにすればよい

```python
from control.matlab import *
```

```python
n = [3]
```

```python
d = [2,1]
```

```python
P = tf(n, d)
```

```python
P
```

```python
t = np.linspace(0, 10, 1000)
```

```python
y, t = step(P, t)
```

```python
plt.plot(t, y)
plt.show()
```
