### 1. Python-control を使い，次式の 2 次システム $P2(s) \omega = 0.2, 1.0, 2.0$ rad/s の正弦波 $sin(ωt)$ を入 力したときの応答をそれぞれ調べよ。プロット範囲は 5 周期分とし，入力と出力を重ねてプロット すること.

```python
# 各種ライブラリインポート
from control.matlab import *
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = [1]
d = [1, 1]

#システムの伝達関数を定義
P = tf(n, d)

w = [0.2, 1.0, 2.0]
```

```python
#グラフスタイル設定
color = ["blue", "green", "red"]
linestyle = ["-", "--", "-."]
```

```python
for i in range(len(w)):
    #横軸の定義
    t = np.linspace(0, 5*2*np.pi/w[i], 1000)
    # 応答を生成
    u = np.sin(w[i]*t)
    y, t, x = lsim(P, u, t)
    #グラフ描画
    plt.plot(t, u, color=color[1], linestyle=linestyle[1],
             label="u, w = {}".format(w[i]))
    plt.plot(t, y, color=color[0], linestyle=linestyle[0],
             label="y, w = {}".format(w[i]))

    plt.ylim(-5.0, 5.0)
    plt.xlabel("t [s]")
    plt.ylabel("y")
    plt.legend()
    plt.grid()
    plt.show()
```

### 2. プロットした結果からそれぞれの周波数におけるゲイン特性と位相特性を考察せよ。大きくなる/ 小さくなる，進む/遅れるといった定性的なものでよい。(必要があればプロット範囲を狭めて拡大 するとよい)
ゲイン特性について
$\omega$を大きくしてにいくと，ゲインは小さくなり，入力に対し出力の振幅は減少する．

位相特性について
$\omega$を大きな値にすると，位相の遅れは大きくなる．
