1. 以下の 2 つの伝達関数についてステップ応答を描画し，その違いを考察せよ(グラフは重ねなくて よい)。$t$ の範囲は 0 s から 5 s までで 100 点取ることととする。
 - $P_1(s)$:零点なし $z$ = []，極 $p$ = −1±5$j$，ゲイン $k$ = 1
 - $P_2(s)$:零点なし $z$ = []，極 $p$ = +1±5$j$，ゲイン $k$ = 1
2. 以下の 3 つの伝達関数についてステップ応答を描画し，その違いを考察せよ(グラフは重ねる)。 $t$ の範囲は 0 s から 5 s までで 100 点取ることととする。
プロットするグラフは以下のような書式にすること
 - $P_3(s)$:零点なし $z$ = []，極 p = −1±5$j$，ゲイン $k$ = 26
 - $P_4(s)$:零点なし $z$ = []，極 p = −3±5$j$，ゲイン $k$ = 34
 - $P_5(s)$:零点なし $z$ = []，極 p = −5±5$j$，ゲイン $k$ = 50

```python
# 各種ライブラリインポート
import numpy as np
import matplotlib.pyplot as plt
from control.matlab import *
```

```python
#横軸の定義
t = np.linspace(0,5,100)
```

 ### 1. 以下の 2 つの伝達関数についてステップ応答を描画し，その違いを考察せよ(グラフは重ねなくて よい)。$t$ の範囲は 0 s から 5 s までで 100 点取ることととする。
 #### $P_1(s)$:零点なし $z$ = []，極 $p$ = −1±5$j$，ゲイン $k$ = 1

```python
z = []
p = [-1+5j, -1-5j]
k = 1

n, d = zpk2tf(z, p, k)
#システムの伝達関数を定義
P = tf(n, d)
#システムのステップ応答を計算
y, t = step(P, t)

plt.plot(t, y, label="z = [], p = [-1+5j, -1-5j], k = 1")

plt.xlabel("t [s]")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

 #### $P_1(s)$:零点なし $z$ = []，極 $p$ = +1±5$j$，ゲイン $k$ = 1

```python
z = []
p = [+1+5j, +1-5j]
k = 1

n, d = zpk2tf(z, p, k)
#システムの伝達関数を定義
P = tf(n, d)
#システムのステップ応答を計算
y, t = step(P, t)

plt.plot(t, y, label="z = [], p = [-1+5j, -1-5j], k = 1")

plt.xlabel("t [s]")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

### 2. 以下の 3 つの伝達関数についてステップ応答を描画し，その違いを考察せよ(グラフは重ねる)。 $t$ の範囲は 0 s から 5 s までで 100 点取ることととする。
 - $P_3(s)$:零点なし $z$ = []，極 p = −1±5$j$，ゲイン $k$ = 26
 - $P_4(s)$:零点なし $z$ = []，極 p = −3±5$j$，ゲイン $k$ = 34
 - $P_5(s)$:零点なし $z$ = []，極 p = −5±5$j$，ゲイン $k$ = 50

```python
z = []
p = [[-1+5j, -1-5j], [-3+5j, -3-5j], [-5+5j, -5-5j]]
k = [26, 34, 50]

#グラフスタイル設定
color = ["blue", "green", "red"]
linestyle = ["-", "--", "-."]

for i in range(len(p)):
    n, d = zpk2tf(z, p[i], k[i])
    #システムの伝達関数を定義
    P = tf(n, d)
    #システムのステップ応答を計算
    y, t = step(P, t)

    plt.plot(t, y, color=color[i], linestyle=linestyle[i],
             label="z = [], p = {}, k = {}".format(p[i], k[i]))

plt.xlabel("t [s]")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
